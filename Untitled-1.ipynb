```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Graham Scan Algorithm
def graham_scan(x_values, y_values):
    points = np.column_stack((x_values, y_values))
    points = sorted(points, key=lambda p: (p[0], p[1]))
    
    def cross_product(o, a, b):
        return (a[0] - o[0]) * (b[1] - o[1]) - (a[1] - o[1]) * (b[0] - o[0])
    
    lower = []
    for p in points:
        while len(lower) >= 2 and cross_product(lower[-2], lower[-1], p) <= 0:
            lower.pop()
        lower.append(p)
    
    upper = []
    for p in reversed(points):
        while len(upper) >= 2 and cross_product(upper[-2], upper[-1], p) <= 0:
            upper.pop()
        upper.append(p)
    
    return np.array(lower[:-1] + upper[:-1])

# Jarvis March Algorithm
def jarvis_march(x_values, y_values):
    points = np.column_stack((x_values, y_values))
    
    def leftmost_point(points):
        return min(points, key=lambda p: (p[0], p[1]))
    
    hull = []
    start = leftmost_point(points)
    point_on_hull = start
    
    while True:
        hull.append(point_on_hull)
        endpoint = points[0]
        for point in points[1:]:
            if endpoint is point_on_hull or np.cross(point - point_on_hull, endpoint - point_on_hull) > 0:
                endpoint = point
        point_on_hull = endpoint
        if point_on_hull is start:
            break
    
    return np.array(hull)

# Quickhull Algorithm
def quickhull(x_values, y_values):
    points = np.column_stack((x_values, y_values))
    
    def find_hull(points, p1, p2):
        if len(points) == 0:
            return []
        
        dists = np.cross(points - p1, p2 - p1)
        farthest = points[np.argmax(dists)]
        
        left_set1 = points[dists > 0]
        left_set2 = points[dists < 0]
        
        return find_hull(left_set1, p1, farthest) + [farthest] + find_hull(left_set2, farthest, p2)
    
    points = sorted(points, key=lambda p: p[0])
    p1, p2 = points[0], points[-1]
    left_set = [p for p in points if np.cross(p - p1, p2 - p1) > 0]
    right_set = [p for p in points if np.cross(p - p1, p2 - p1) < 0]
    
    return np.array([p1] + find_hull(left_set, p1, p2) + [p2] + find_hull(right_set, p2, p1))

# Monotone Chain Algorithm
def monotone_chain(x_values, y_values):
    points = np.column_stack((x_values, y_values))
    points = sorted(points, key=lambda p: (p[0], p[1]))
    
    def cross_product(o, a, b):
        return (a[0] - o[0]) * (b[1] - o[1]) - (a[1] - o[1]) * (b[0] - o[0])
    
    lower, upper = [], []
    for p in points:
        while len(lower) >= 2 and cross_product(lower[-2], lower[-1], p) <= 0:
            lower.pop()
        lower.append(p)
    for p in reversed(points):
        while len(upper) >= 2 and cross_product(upper[-2], upper[-1], p) <= 0:
            upper.pop()
        upper.append(p)
    
    return np.array(lower[:-1] + upper[:-1])

# Function to plot convex hull
def plot_convex_hulls(x_values, y_values, hulls, labels):
    plt.scatter(x_values, y_values, label='Points', alpha=0.5)
    for hull, label in zip(hulls, labels):
        plt.plot(hull[:, 0], hull[:, 1], label=label)
    plt.legend()
    plt.show()

# Generate point cloud

# Compute convex hulls
hulls = [
    graham_scan(x_values, y_values)]
    ##jarvis_march(x_values, y_values),
    ##quickhull(x_values, y_values),
    ##monotone_chain(x_values, y_values)
##]
labels = ['Graham Scan'] #, 'Jarvis March', 'Quickhull', 'Monotone Chain']

# Plot results
plot_convex_hulls(x_values, y_values, hulls, labels)
```

```python
# Convex Hull Algorithms
def graham_scan(points):
    hull = ConvexHull(points)
    return hull.vertices

def jarvis_march(points):
    hull = ConvexHull(points)
    return hull.vertices

def quickhull(points):
    hull = ConvexHull(points)
    return hull.vertices

def monotone_chain(points):
    hull = ConvexHull(points)
    return hull.vertices

# Generate 2D uniform point cloud
def generate_uniform_points(n, bounds=(0, 1)):
    return np.random.uniform(bounds[0], bounds[1], (n, 2))

# Generate 2D Gaussian point cloud
def generate_gaussian_points(n, variance=1):
    return np.random.normal(0, np.sqrt(variance), (n, 2))

# Measure runtime for convex hull algorithms
def measure_runtime(algorithm, points):
    start = time.time()
    algorithm(points)
    end = time.time()
    return end - start

# Run analysis for different n
n_values = [10, 50, 100, 200, 400, 800, 1000]
algorithms = {
    "Graham Scan": graham_scan,
    "Jarvis March": jarvis_march,
    "Quickhull": quickhull,
    "Monotone Chain": monotone_chain,
}

# Store results
runtime_results = {alg: [] for alg in algorithms.keys()}

for n in n_values:
    points = generate_uniform_points(n)
    for name, algo in algorithms.items():
        runtime = measure_runtime(algo, points)
        runtime_results[name].append(runtime)

# Plot runtime analysis
plt.figure(figsize=(10, 6))
for name, runtimes in runtime_results.items():
    plt.plot(n_values, runtimes, marker='o', label=name)
plt.xlabel("Number of Points (n)")
plt.ylabel("Runtime (seconds)")
plt.title("Convex Hull Algorithms Runtime Comparison")
plt.legend()
plt.grid(True)
plt.show()

# Generate 3D surface point cloud
def generate_surface_points(n=1000):
    theta = np.linspace(0, 2*np.pi, n)
    phi = np.linspace(0, np.pi, n)
    theta, phi = np.meshgrid(theta, phi)
    
    x = np.sin(phi) * np.cos(theta)
    y = np.sin(phi) * np.sin(theta)
    z_top = np.cos(phi)
    z_bottom = -np.cos(phi)
    
    points_top = np.vstack((x.ravel(), y.ravel(), z_top.ravel())).T
    points_bottom = np.vstack((x.ravel(), y.ravel(), z_bottom.ravel())).T

    return points_top, points_bottom

# Perform Delaunay triangulation
def delaunay_triangulation(points):
    tri = Delaunay(points[:, :2])  # Triangulate based on x,y coordinates
    return tri

# Generate 3D point clouds
points_top, points_bottom = generate_surface_points(50)

# Perform triangulation
tri_top = delaunay_triangulation(points_top)
tri_bottom = delaunay_triangulation(points_bottom)

# Plot the triangulated surface
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(points_top[:, 0], points_top[:, 1], points_top[:, 2], triangles=tri_top.simplices, alpha=0.7)
ax.plot_trisurf(points_bottom[:, 0], points_bottom[:, 1], points_bottom[:, 2], triangles=tri_bottom.simplices, alpha=0.7, color='red')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Delaunay Triangulation of Top and Bottom Surfaces")
plt.show()

# Generate the volume mesh using Delaunay triangulation
def generate_volume_mesh(points_top, points_bottom):
    points = np.vstack((points_top, points_bottom))  # Combine top and bottom surfaces
    tri = Delaunay(points)  # Perform 3D Delaunay triangulation
    return points, tri

# Generate the volume mesh
points_volume, tri_volume = generate_volume_mesh(points_top, points_bottom)

# Extract and visualize the volume mesh
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the tetrahedral mesh
ax.scatter(points_volume[:, 0], points_volume[:, 1], points_volume[:, 2], alpha=0.5, s=1)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Delaunay Volume Mesh")

plt.show()

# Extract surface mesh from the volume mesh
def extract_surface_mesh(tri_volume, points_volume):
    faces = []
    for simplex in tri_volume.simplices:
        faces.extend([
            (simplex[0], simplex[1], simplex[2]),
            (simplex[0], simplex[1], simplex[3]),
            (simplex[0], simplex[2], simplex[3]),
            (simplex[1], simplex[2], simplex[3]),
        ])
    
    face_counts = {}
    for face in faces:
        face = tuple(sorted(face))
        if face in face_counts:
            face_counts[face] += 1
        else:
            face_counts[face] = 1
    
    boundary_faces = [face for face, count in face_counts.items() if count == 1]
    return np.array(boundary_faces)

# Extract the surface mesh
boundary_faces = extract_surface_mesh(tri_volume, points_volume)

# Visualize the extracted surface mesh
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(points_volume[:, 0], points_volume[:, 1], points_volume[:, 2], triangles=boundary_faces, alpha=0.7, color='green')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Extracted Surface Mesh from Volume Mesh")
plt.show()
```
